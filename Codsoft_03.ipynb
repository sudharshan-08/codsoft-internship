```python
lab=LabelEncoder()
risk['species']=lab.fit_transform(risk['species'])
```

```python

risk.head()
```

```python


x=risk.drop(columns='species')
```

```python
y=risk['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python

log=LogisticRegression()
```

```python


log.fit(x_train,y_train)
```

```python
LogisticRegression()
```

```python

knc=KNeighborsClassifier()
```

```python


knc.fit(x_train,y_train)
```

```python
KNeighborsClassifier()
```

```python
dtc=DecisionTreeClassifier()
```

```python

dtc.fit(x_train,y_train)
```

```python

DecisionTreeClassifier()
```

```python


log_accuracy=log.score(x_test,y_test)*100
knc_accuracy=knc.score(x_test,y_test)*100
dtc_accuracy=dtc.score(x_test,y_test)*100
```

```python


print(f"Accuracy by using Logistic Regression: {log_accuracy}%")
```

```python

print(f"Accuracy by using K Nearest Neighbors Algorithm: {knc_accuracy}%")
```

```python
print(f"Accuracy by using Decision Tree Classifier: {dtc_accuracy}%")
```
